```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline  
#%matplotlib qt
# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
#Path a imagenes
img_path = "TP1"
```

```python
# Cargo imagen 1
path = img_path + '\img1_tp.png'
img_tp_1 = cv.imread(path, cv.IMREAD_GRAYSCALE)

plt.figure()
plt.imshow(img_tp_1, cmap='gray')
plt.show()
```

```python
# Cargo imagen 2
path = img_path + '\img2_tp.png'
img_tp_2 = cv.imread(path, cv.IMREAD_GRAYSCALE)

plt.figure()
plt.imshow(img_tp_2, cmap='gray')
plt.show()
```

```python
#Armo los histogramas
hist1,bins1 = np.histogram(img_tp_1.ravel(),100,[0,256])
hist2,bins2 = np.histogram(img_tp_2.ravel(),100,[0,256])

fig = plt.figure()
ax1=plt.subplot(221)
ax1.plot(hist1)
ax1.set_title('IM1')

ax2=plt.subplot(222)
ax2.plot(hist2)
ax2.set_title('IM2')
```

En ambos casos el histograma es muy parecido, por lo que le sería muy complejo a un modelo de detección emplearlo para reconocer algo

```python
# Cargo imagen 2
path = img_path + '\segmentacion.png'
img_seg = cv.imread(path, cv.IMREAD_COLOR)
img_seg = cv.cvtColor(img_seg, cv.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img_seg)
plt.show()
```

```python
#Armo los histogramas
hist1,bins1 = np.histogram(img_seg[:,:,0].ravel(),256,[0,256])
hist2,bins2 = np.histogram(img_seg[:,:,1].ravel(),256,[0,256])
hist3,bins3 = np.histogram(img_seg[:,:,2].ravel(),256,[0,256])

fig = plt.figure()
ax1=plt.subplot(311)
ax1.plot(hist1)
ax1.set_title('H1')

ax2=plt.subplot(312)
ax2.plot(hist2)
ax2.set_title('H2')

ax3=plt.subplot(313)
ax3.plot(hist3)
ax3.set_title('H3')
```

```python
#Para verde
color_l = (0,50,150)
color_u = (255,200,255)
mask = cv.inRange(img_seg, color_l,  color_u)
img_segmentada1 = cv.bitwise_and(img_seg, img_seg, mask=mask)

#Para azul
color_l = (0,0,50)
color_u = (30,100,150)
mask = cv.inRange(img_seg, color_l,  color_u)
img_segmentada2 = cv.bitwise_and(img_seg, img_seg, mask=mask)

#Para rojo
color_l = (0,0,0)
color_u = (255,255,50)
mask = cv.inRange(img_seg, color_l,  color_u)
img_segmentada3 = cv.bitwise_and(img_seg, img_seg, mask=mask)


fig = plt.figure()

ax1=plt.subplot(221)
ax1=plt.imshow(img_segmentada1)

ax2=plt.subplot(222)
ax2=plt.imshow(img_segmentada2)

ax3=plt.subplot(223)
ax3=plt.imshow(img_segmentada3)

ax4=plt.subplot(224)
ax4=plt.imshow(img_seg)
```
